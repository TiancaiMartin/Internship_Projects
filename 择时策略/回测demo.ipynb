```python
#行情数据
sec_post_1d_df = pd.read_pickle(r'D:\coding\data\cta\basic_info\k_bar\continuous\dom_cont_post_1d')
sec_post_1d_df
```

```python
#待测信号 这里的信号为当日收盘后可得的信号。第二个交易根据信号进行交易。例如双均线策略，当日ma(close,5)>ma(close,20),为1；反之为-1.
test_sig_df = pd.read_pickle(r'D:\software\WeChat\聊天记录\WeChat Files\xifengshouma038325\FileStorage\File\2023-12\tech_ts_risk')
test_sig_df
```

```python
#收益率序列
pct_df = sec_post_1d_df.xs('close',level=1).pct_change(axis=1).reindex(test_sig_df.index)
```

```python
#手续费
fee_df = abs(test_sig_df.diff(axis=1).fillna(0))*1e-4 #单边万一的手续费
```

```python
#当天的信号作用到第二天，用close计算出来的收益率，shift-1.  

pnl_seires = ((pct_df.shift(-1,axis=1).fillna(0)*test_sig_df-fee_df).mean()+1).cumprod()
```
