```python
import pandas as pd
from openpyxl import load_workbook
import numpy as np
# 读取Excel文件
wb = load_workbook('Data_Test.xlsx', data_only=True)
ws = wb['Previous']
data = pd.DataFrame(ws.values)
```

```python
data = data.iloc[1:].reset_index(drop=True)
data
```

```python
data.columns = data.iloc[0]
data = data.iloc[1:].reset_index(drop=True)
data.columns = ['Date'] + list(data.columns[1:])
data = data.set_index('Date').sort_index()
data.replace(0, np.nan, inplace=True)       # 数值 0 → NaN
data.replace(['0', '0.0'], np.nan, inplace=True)   # 字符串 '0' 或 '0.0' → NaN
data
```

```python
# 合并所有映射关系
rename_map = {
    "中国:CPI:当月同比": "CPI_同比",
    "中国:CPI:不包括食品和能源(核心CPI):当月同比": "核心CPI_同比",
    "中债浮动利率企业债(Depo-1Y)点差曲线(AAA):1年": "AAA_公司债利率",  # 中国没有BAA，所以用AAA企业债利率代替
    "中国:固定利率国债:发行利率:10年": "10Y国债利率",
    "中国:固定利率国债:发行利率:3个月": "3M_TBill利率",
    "中国:逆回购利率:7天": "政策利率",
    "中债固定利率国债全价指数": "IEF",
    "中国:固定利率国债:发行利率:2年": "2Y国债利率",
    "中国:制造业PMI": "PMI",
    "中国:GDP:现价:当季值": "GDP_现价",
    "期货收盘价(活跃合约):INE原油": "WTI",
    "期货结算价(活跃合约):黄金": "GLD",
    "沪深300指数": "HS300",
}
data = data.rename(columns=rename_map)
```

```python
data
```

```python
data_monthly = data.resample('ME').last()
# 使用线形插值法填充数据的缺失值
data_monthly = data_monthly.interpolate(method='linear')
data_monthly
```

```python
# 绘制清晰好的数据的图表
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Songti SC']  # 中文支持
plt.rcParams['axes.unicode_minus'] = False       # 负号显示
data_monthly[['CPI_同比', '10Y国债利率', '3M_TBill利率', '政策利率']].plot(figsize=(12, 8))
plt.xlabel('日期')
plt.ylabel('指标值')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# 导出为csv文件
data_monthly.to_csv('data_China.csv')
```
